```python
import pandas as pd

# List of files to concatenate
file_paths = [
    "/content/drive/MyDrive/Colab Notebooks/UNSW_2018_IoT_Botnet_Full5pc_1.csv",
    "/content/drive/MyDrive/Colab Notebooks/UNSW_2018_IoT_Botnet_Full5pc_2.csv",
    "/content/drive/MyDrive/Colab Notebooks/UNSW_2018_IoT_Botnet_Full5pc_3.csv",
    "/content/drive/MyDrive/Colab Notebooks/UNSW_2018_IoT_Botnet_Full5pc_4.csv"
]

# Taking exactly 25,000 records from each file
sample_size = 25000  # Exact number of rows to take from each file
dataframes = []

for file in file_paths:
    # Read the file
    df = pd.read_csv(file)
    print(f"Original file {file}, shape: {df.shape}")

    # Sample exactly 25,000 records
    # If file has fewer than 25,000 records, it will take all available records
    if len(df) > sample_size:
        sampled_df = df.sample(n=sample_size, random_state=42)
    else:
        sampled_df = df
        print(f"Warning: {file} has fewer than {sample_size} records, taking all available")

    print(f"After sampling {sample_size} records: {sampled_df.shape}")
    dataframes.append(sampled_df)

# Combine all sampled dataframes
combined_df = pd.concat(dataframes, ignore_index=True)

# Show combined results
print(f"\nCombined dataframe shape: {combined_df.shape}")

# Save to a new file
output_path = "/content/drive/MyDrive/Colab Notebooks/UNSW_2018_IoT_Botnet_25K_Each.csv"
combined_df.to_csv(output_path, index=False)
print(f"Trimmed data saved to {output_path}")
```
